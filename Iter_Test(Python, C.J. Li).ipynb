```python
import numpy as np
import time
import matplotlib.pyplot as plt

iter = 100
time_1 = 0

#设置列表储存变量
X = []
for i in range(iter):
    start_time = time.time()
    n = 2500
    A = np.random.rand(n, n)
    trans_A = A.T
    A_Square = A * A
    time_1 += time.time() - start_time
    #清空变量
    del A
    #计算每一次的平均时间，并插入列表
    time_0 = time_1/(i+1)
    X.append(time_0)

time_1 /= iter

plt.plot(range(iter),X)
```
